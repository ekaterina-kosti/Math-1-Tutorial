## Exercise 2: Integration by part

a) $\int x^2 \ln(x) \, dx$ 

b) $\int \sin(\ln(x)) \, dx$

Integration by Part

Step 1: Choose the first and the second function according to the ILATE rule. 

Step 2: Differentiate u(x) with respect to x that is, Evaluate du/dx.

Step 3: Integrate v(x) with respect to x that is, Evaluate ‚à´v dx.

The formula for Integration by part,

$\int uv \, dx = u\int v \, dx ‚àí \int ((du/dx)\int v \, dx) \, dx$

```python
# --- Exercise 2: Integration by parts ---

print("2a:", sp.integrate(x**2 * sp.ln(x), x) + C)
print("2b:", sp.integrate(sp.sin(sp.ln(x)), x) + C)
```

## Exercise 3: Trigonometric substitution

a) $\int \frac{1}{\sqrt{16-9x^2}} \, dx$

b) $\int \frac{1}{9x^2+16} \, dx$

c) $\int \frac{1}{x^2+6x+25} \, dx$

d) $\int e^x \sqrt{1-e^{2x}} \, dx$

Here we use the forms:

$\int \frac{1}{\sqrt{x^2 + a^2}} \, dx$ = $\frac{1}{a} (\tan(x/a))^{-1}$ and
$\int \frac{1}{\sqrt{a^2 - a^2}} \, dx$ = $(\sin(x/a))^{-1}$

```python
# --- Exercise 3: Trigonometric substitution ---

print("3a:", sp.integrate(1 / sp.sqrt(16 - 9*x**2), x) + C)
print("3b:", sp.integrate(1 / (9*x**2 + 16), x) + C)
print("3c:", sp.integrate(1 / (x**2 + 6*x + 25), x) + C)
print("3d:", sp.integrate(sp.exp(x) * sp.sqrt(1 - sp.exp(2*x)), x) + C)
```

## Exercise 4: Partial fractions

a) $\int \frac{9x+32}{x^2+6x+8} \, dx$ (linear factors)

b) $\int \frac{4}{(x-1)(x+1)^2} \, dx$ (repeated factors)

c) $\int \frac{2x^2-3x+22}{(x-4)(x^2+5)} \, dx$ (irreducible quadratic factor)

```python
# --- Exercise 4: Partial fractions ---

print("4a:", sp.integrate((9*x + 32) / (x**2 + 6*x + 8), x) + C)
print("4b:", sp.integrate(4 / ((x - 1) * (x + 1)**2), x) + C)
print("4c:", sp.integrate((2*x**2 - 3*x + 22) / ((x - 4) * (x**2 + 5)), x) + C)
```

## Exercise 5: Trigonometric & Hyperbolic Integrals

a) $\int \frac{dx}{1-\cos x}$
*Hint: use the half-angle identity*

b) $\int \tan^5 x \, dx$
*Hint: use the (secx)^2 identity. You will also have to derive tan'(x) = -ln(cos(x))*

c) $\int \cosh^2 x \sinh x \, dx$

d) $\int \frac{\sinh x}{1+\cosh x} \, dx$

e) $\int \frac{dx}{\sqrt{4+x^2}}$
*Hint: use trigonometric substitution*

```python
# --- Exercise 5: Trigonometric & Hyperbolic Integrals ---

print("5a:", sp.integrate(1 / (1 - sp.cos(x)), x) + C)
print("5b:", sp.integrate(sp.tan(x)**5, x) + C)
print("5c:", sp.integrate(sp.cosh(x)**2 * sp.sinh(x), x) + C)
print("5d:", sp.integrate(sp.sinh(x) / (1 + sp.cosh(x)), x) + C)
print("5e:", sp.integrate(1 / sp.sqrt(4 + x**2), x) + C)
```

```python
# Define variables
x = sp.symbols('x')

def solve_and_print(title, problems):
    print(f"--- {title} ---")
    for label, expr in problems.items():
        result = sp.integrate(expr, x)
        display(sp.Eq(sp.Integral(expr, x), result + sp.symbols('C')))

# Exercise 1
ex1 = {
    "1a": x * sp.cos(x**2),
    "1b": sp.exp(x) / sp.sqrt(1 + sp.exp(x)),
    "1c": sp.ln(x) / x
}

# Exercise 2
ex2 = {
    "2a": x**2 * sp.ln(x),
    "2b": sp.sin(sp.ln(x))
}

# Exercise 3
ex3 = {
    "3a": 1 / sp.sqrt(16 - 9*x**2),
    "3b": 1 / sp.sqrt(9*x**2 + 16),
    "3c": 1 / (x**2 + 6*x + 25),
    "3d": sp.exp(x) * sp.sqrt(1 - sp.exp(2*x))
}

# Exercise 4
ex4 = {
    "4a": (9*x + 32) / (x**2 + 6*x + 8),
    "4b": 4 / ((x - 1) * (x + 1)**2),
    "4c": (2*x**2 - 3*x + 22) / ((x - 4) * (x**2 + 5))
}

# Exercise 5
ex5 = {
    "5a": 1 / (1 - sp.cos(x)),
    "5b": sp.tan(x)**5,
    "5c": sp.cosh(x)**2 * sp.sinh(x),
    "5d": sp.sinh(x) / (1 + sp.cosh(x)),
    "5e": 1 / sp.sqrt(4 - x**2)
}

# Run solutions
solve_and_print("Exercise 1", ex1)
solve_and_print("Exercise 2", ex2)
solve_and_print("Exercise 3", ex3)
solve_and_print("Exercise 4", ex4)
solve_and_print("Exercise 5", ex5)
```
