### Exercise 8: Taylor Series

Taylor Series of Logarithmic Functions

a) *Taylor Series Expansion* Find the Taylor series of $f(x) = \ln(1 + x)$ centered at $a = 0$ (also known as the Maclaurin series).

b) *Convergence* Determine the radius of convergence $R$ of this Taylor series.

c) *Approximation* Use the first two non-zero terms of the power series found in (a) to approximate $\ln(3/2)$.

```python


# (a) Find Taylor series centered at a=0 up to x^4
# The series for ln(1+x) is x - x^2/2 + x^3/3 - x^4/4 + ...
f = sp.log(1 + x)
taylor_series = sp.series(f, x, 0, 5)
print(f"Taylor series: {taylor_series}")

# (b) Radius of Convergence
R = 1 
print(f"Radius of Convergence R: {R}")

# (c) Approximate ln(3/2) 
# ln(3/2) = ln(1 + 1/)
x_val = 0.5
# First two non-zero terms: x - x^2/2
approx_expr = x - (x**2)/2
approximation = approx_expr.subs(x, x_val)
exact_val = sp.log(1.5).evalf()

print(f"Approximation of ln(1.5): {approximation}")
print(f"Exact value (float): {exact_val}")


```
