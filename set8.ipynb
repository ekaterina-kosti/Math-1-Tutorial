### Exercise 2: Limits
Calculate the following limits:
* **a)** $\lim_{h \to 0} \frac{\sqrt{5h + 4} - 2}{h}$
* **b)** $\lim_{x \to 2} \frac{x + 3}{x + 6}$
* **c)** $\lim_{x \to -2} \frac{x + 2}{\sqrt{x^2 + 5} - 3}$

```python
# a)
f_a = (sp.sqrt(5*x + 4) - 2 )/ x
print(f"The limit of a) as x approaches 0 is ", sp.limit(f_a, x, 0))
# b)
f_b = (x+3) / (x+6)
print(f"The limit of b) as x approaches 2 is ", sp.limit(f_b, x, 2))
#c)
f_c = (x+2) / (sp.sqrt(x**2 +5) - 3)
print(f"The limit of c) as x approaches -2 is ", sp.limit(f_c, x, -2))
```

### Exercise 5: Physics/Constants
Given $P = \frac{nRT}{V - nb} - \frac{an^2}{V^2}$ where $a, b, n, R$ are constants, find $\frac{dP}{dV}$.

```python
V = sp.symbols('V')
T = sp.symbols('T')
a, b, n, R = sp.symbols('a, b, n, R') # define constants 
sp.diff((n*R*T) / (V - n*b) - (a*n**2) / (V**2), V)
```

### Exercise 6: Derivatives
#### Trigonometric
Find the derivatives of:
* **a)** $f(x) = \sin x \tan x$
* **b)** $y = x^2 \cot x - \frac{1}{x^2}$

```python
# a)
sp.expand(sp.simplify(sp.diff(sp.sin(x)*sp.tan(x)), Trig=True))
```

```python
# b)
sp.expand(sp.simplify(sp.diff((sp.sin(x) + sp.cos(x))*sp.sec(x)), Trig=True))
```

```python
# c)
sp.expand(sp.simplify(sp.diff(x**2*sp.cot(x) - 1/x**2), Trig=True))
```

#### Implicit Differentiation
Use implicit differentiation:
* **a)** $y \sin(xy) = y^2 + 2$
* **b)** $x = \cos(xy)$
* **c)** $x = \tan(y)$

```python
y = sp.symbols('y')
# a)
sp.idiff(y*sp.sin(x*y) - y**2 - 2,y,x)
```

```python
# b)
sp.idiff(x - sp.cos(x*y),y,x)
```

```python
# c)
sp.idiff(x - sp.tan(y), y, x)
```
