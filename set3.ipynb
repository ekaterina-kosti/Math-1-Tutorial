Inverse matrices and the detrminant

```python
import numpy as np
import numpy.linalg as la
```

Exercise 1.

a) show that det(A²) = det(A)²
b) show that B = A-1. Use A.A-1=I
c) Calculate det(A) and det (B) and show that
det(A-1) = 1/det(A)

```python
matrix_A = np.array([[1, 3, 0], [2, 4, 0], [0, 0, 5]])
matrix_B = np.array([[-2, 1.5, 0], [1, -0.5, 0], [0, 0, 0.2]])
detA_2 = la.det(matrix_A@matrix_A)
print(f"Determinant of A^2: {detA_2:.2f}")
detA = la.det(matrix_A)
detB = la.det(matrix_B)
```
