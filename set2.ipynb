```python
import numpy as np
import numpy.linalg as la # import the necessary modules
```

1.3 & 1.4 - matrix sums and products

```python
# define the matrices
matrix_B = np.array([[3,0],[-1,5]])
matrix_A = np.array([[1, -3], [4, 2]])
matrix_C = np.array([[5],[9]])
matrix_D = np.array([[1,1,3],[4,-2,3]])
matrix_E = np.array([[1, 2],[0,6],[-2,3]])
matrix_G = np.array([[4, 2, 3],[-2, 6, 0],[0, 7, 1]])
matrix_H = np.array([[1, 0, 1],[5, 2, -2],[2, -3, 4]])

sum_e = 3*matrix_B - matrix_A
print(sum_e, "\n\n")
sum_T = matrix_A.T + matrix_B.T
print(sum_T, "\n\n")
sum_T2 = (matrix_A + matrix_B).T
print(f"sum of the transposes is equal to the transpose of the sum", sum_T2, "\n\n")

product_AC = np.matmul(matrix_A,matrix_C)
print(product_AC, "\n\n")
product_INcorrect = matrix_A*matrix_C
print(f"incorrect product calculation with asterisk", product_INcorrect, "\n")

try:
    product_CA = np.matmul(matrix_C, matrix_A)
    print(product_CA)
except ValueError:
    print("Skipping matrix_C @ matrix_A — incompatible dimensions")
```

```python
product_CTC = np.matmul(matrix_C.T,matrix_C)
print(product_CTC, "\n\n")
product_CCT = np.matmul(matrix_C,matrix_C.T)
print(product_CCT, "\n\n")
product_DE = np.matmul(matrix_D,matrix_E)
print(product_DE, "\n\n")
product_GH = np.matmul(matrix_G,matrix_H)
print(product_GH)
```

1.5 Calculate the determinants

```python
detA = la.det(matrix_A)

# detB = la.det(matrix_B)
detG = la.det(matrix_G)
detH = la.det(matrix_H)
print(detA, "\n", detB, "\n", detG, "\n", detH)
```

1.6.A - Calculate the inverse of matrix A

```python
n = matrix_A.shape[0]
minors = np.zeros((n,n))
cofactors = np.zeros((n,n))
for i in range(n):                  # 0-based indices: 0..n-1
    for j in range(n):
        A_sub = np.delete(np.delete(matrix_A, i, axis=0), j, axis=1)
        minor = la.det(A_sub)
        cofactor = ((-1)**(i + j)) * minor
        minors[i, j] = minor
        cofactors[i, j] = cofactor
        print(f"i={i}, j={j} minor: {minor}, cofactor: {cofactor}\n")
# The inverse via cofactors:
if abs(detA) > 0:
    adj = cofactors.T
    A_inv_from_adj = adj / detA
print("Adjugate of A:\n", adj)
print("Inverse from adjugate:\n", A_inv_from_adj)
```

```python
# or 
inverse_A = la.inv(matrix_A)
print("Inverse from linalg.inv:\n", inverse_A)
```

1.6.G - Calculate the inverse of matrix G

```python
n = matrix_G.shape[0]
minors = np.zeros((n,n))
cofactors = np.zeros((n,n))
for i in range(n):                  # 0-based indices: 0..n-1
    for j in range(n):
        G_sub = np.delete(np.delete(matrix_G, i, axis=0), j, axis=1)
        minor = la.det(G_sub)
        cofactor = ((-1)**(i + j)) * minor
        minors[i, j] = minor
        cofactors[i, j] = cofactor
        print(f"i={i}, j={j} minor: {minor}, cofactor: {cofactor}\n")
# The inverse via cofactors:
if abs(detG) > 0:
    adj = cofactors.T
    G_inv_from_adj = adj / detG
print("Adjugate of G:\n", adj)
print("Inverse from adjugate:\n", G_inv_from_adj)
```

```python
# or 
inverse_G = la.inv(matrix_G)
print("Inverse from linalg.inv:\n", inverse_G)
```
