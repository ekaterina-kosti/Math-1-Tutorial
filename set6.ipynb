## Exercise 1: Domain and Range of Functions
Graph the functions. 

function a) is discontinuous on t = +- 4. To plot the discontinuity we can define a function `discont(x)`

```python
# --- N1: Function Plotting ---

# N1. a) f(t) = 2 / (t^2 - 16)
x = np.linspace(-5, 5, 100)

def discont(x):
    """See http://stackoverflow.com/a/15122658/554319"""
    y = 2 / (x**2 - 16)
    mask = (np.abs(x - 4) < 0.03) | (np.abs(x + 4) < 0.03)
    y[mask] = np.nan
    return y

plt.plot(x, discont(x), color='blue')

plt.axhline(0, color='black', linewidth=1)  # x-axis
plt.axvline(0, color='black', linewidth=1)  # y-axis

plt.ylim(-4, 4)
plt.xlim(-5, 5)
plt.grid(True)
plt.title("a) $f(t) = \\frac{2}{t^2 - 16}$")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.show()
```

## Exercise 2: Find the Period of Each Function

The period of $\text{trig}(Bx)$ is $P = \text{Base Period}/|B|$

Graph the functions

Determine the symmetry

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle(r'N2: Trig Function Graphs ', fontsize=16)

# N2. a) sin(2x)+1
x_a = np.linspace(-10, 10, 500)
g_a = np.sin(2*x_a) + 1
axes[0,0].plot(x_a, g_a)
axes[0, 0].set_title(r'$\sin(2x)+1$')
axes[0, 0].set_ylim(-1, 3)

# N2. b) cos(x-pi/2)
x_b = np.linspace(-10, 10, 500)
g_b = np.cos(x_b - np.pi/2)
axes[0,1].plot(x_b, g_b)
axes[0, 1].set_title(r'$\cos(x-\pi/2)$')
axes[0, 1].set_ylim(-1, 3)

# N2. c) -tan(pi*t)
x_c = np.linspace(-10, 10, 500)
g_c = -np.tan(np.pi*x_c)
axes[1,0].plot(x_c, g_c)
axes[1, 0].set_title(r'$-tan(\pi *t)$')
axes[1, 0].set_ylim(-5, 5)

# N2. d) cot(x) + 3
x_d = np.linspace(-10, 10, 500)
g_d = 1/ np.tan(x_d) +3 
axes[1,1].plot(x_d, g_d)
axes[1, 1].set_title(r'$\cot(x)+3$')
axes[1, 1].set_ylim(-5, 5)
```

## Exercise 3: Solve the trignometric equations 

for theta in [0, 2pi]

Use `solveset` from `sympy` package.

```python
# --- N3: Solving Trigonometric Equations ---

# Define the symbols
x = symbols('x', real=True)
theta = symbols('theta', real=True)

# The interval for the solutions [0, 2pi)]
interval = Interval(0, 2*pi, right_open=True)


# 1. N3. a) sin^2(theta) = 3/4 

sol_a = solveset(sin(theta)**2 -3/4, domain=interval)

# 2. N3. b) sin(2*theta) = cos(theta)
sol_b = solveset(sin(2*theta) - cos(theta), domain=interval)

# 3. N3. c) tan(theta - pi/2) = 1

sol_c = solveset(tan(theta - pi/2) -1)
# 4. N3. d) cos(3x + 2) = 1/sqrt(2)
sol_d = solveset(cos(3*x + 2) - 1/sp.sqrt(2))


print(f"N3.a) Equation sin^2(theta) = 3/4: Solutions are {sol_a}\n")
print(f"N3.b) Equation sin(2*theta) = cos(theta): Solutions are {sol_b}\n")
print(f"N3.c) Equation tan(theta - pi/2) = 1: Solutions are {sol_c}\n")
print(f"N3.d) Equation cos(3x + 2) = 1/sqrt(2): Solutions are\n {sol_d}\n")
```

## Exercise 6: Polar Curve


Sketch in polar coordinates r = 1 + $\sin(\theta)$

Find cartesian parametric and rectangular description.

```python
# --- N6: Polar Plotting (r = 1 + sin(theta)) ---
theta = np.linspace(0, 2 * np.pi, 500)
r = 1 + np.sin(theta)

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, r, color='blue')
ax.set_title('Polar Curve $r = 1 + \\sin\\theta$', va='bottom')
# plt.savefig('N6_Cardioid_Plot.png')
plt.show()
```
